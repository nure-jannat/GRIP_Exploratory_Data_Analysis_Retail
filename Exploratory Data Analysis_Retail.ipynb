```python
sns.heatmap(df.corr(), annot=True)
```

The correlation between Profit and Sales are 0.48

```python
sns.countplot(data=df, x='Segment', palette='rocket')
plt.title("Segments of Business")
plt.show()
```

```python
sns.barplot(x='Category', y='Sales',data=df)
plt.title("Sales in different Categories of Business")
plt.show()
```

The demand for technology product is high.

```python
sns.barplot(x='Category', y='Profit',data=df, palette='rocket')
plt.title("Profit in different Categories of Business")
plt.show()
```

The profit from furniture product are poor!

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Sub-Category', y='Profit',data=df)
plt.title("Profit in different Sub-Categories")
plt.show()
```

For Bookcases and Tables, profit is less than 0 which indicates loss in business for these 2 product.

```python
sns.barplot(x='Region', y='Profit',hue='Category',data=df, palette='Set1')
plt.title("Profit in different Region according to Categories")
plt.show()
```

Profit is high for technology in East, for Furniture in South and for office supplies in West region.

```python
sns.barplot(x='Region', y='Sales',hue='Category',data=df, palette='Set2')
plt.title("Sales in different Region according to Categories")
plt.show()
```

The demand for technical product are high in South and East region.
The demand for furniture productand office supplies are almost same in each region.

```python
sns.lineplot(x='Discount', y='Sales', data=df, color='g', label='Discount')
plt.legend()
```

Sales for product increases with the discount

```python
sns.lineplot(x='Discount', y='Profit', data=df, color='b', label='Discount')
plt.legend()
```

Profit are less with the increasing rate of discount for product. For maximizin the profit, discount should be in minimum rate.

```python
plt.figure(figsize=(20,5))
sns.barplot(x='State', y='Sales',data=df)
plt.title("Sales in different State")
plt.xticks(rotation=90)
plt.show()
```

The highest selling area is Wyoming.

```python
plt.figure(figsize=(20,5))
sns.barplot(x='State', y='Profit',data=df)
plt.title("Profit in different State")
plt.xticks(rotation=90)
plt.show()
```

The states where loss occurs in business are Florida, North Carolina, Texas, Pennsylvania, Illinois, Arizona, Tennessee, Oregon, Colorado, Ohio. The most profitable state is Vermont.
